```python
url = "https://riskzone.anywhereportfolio.com.ar:9099/api/cuentas/getcuentacompensacion"
df = requests.get(url, verify=False)
data_cuentas  = (df.json())
data_cuentas = pd.DataFrame(data_cuentas)

def get_tipo_de_cuenta(CIMCODIGO):
    CIMCODIGO = str(CIMCODIGO)
    data_mc = ((data_cuentas[data_cuentas["CuentaCompensacionCodigo"] == CIMCODIGO]))
    tipocuenta = data_mc["TipoCuentaCompensacionDescripcion"].iloc[0] #.to_string(Name=False,dtype=False,index=False)
    return(tipocuenta)

def get_mc_name(MC):
    MC = str(MC)
    data_mc = ((data_cuentas[data_cuentas["MiembroCompensadorCodigo"] == MC]))
    name_mc = data_mc["MiembroCompensadorDescripcion"].iloc[0] #.to_string(Name=False,dtype=False,index=False)
    return(name_mc)
```

```python
propia = []
terceros =[]
cer = []
cel=[]
for i in range(len(dfc)):
    saldopropia = dfc.iloc[i,1]
    saldoterceros = dfc.iloc[i,2]
    saldocer = dfc.iloc[i,3]
    saldocel = dfc.iloc[i,4]

    if saldopropia > 0:
        if saldopropia > abs(saldoterceros):
            saldopropia = saldopropia + saldoterceros # aca en realidad se resta, dado que saldo terceros tiene signo negativo
            saldoterceros = 0 # saldo de terceros pasa a estar cubierto y el saldo de la propia pasa a ser lo que le queda despues de cubrir

        if saldopropia > abs(saldocer):
            saldopropia = saldopropia + saldocer
            saldocer = 0
    if saldocel > 0:
        saldocel = 0

        
    propia.append(saldopropia)
    terceros.append(saldoterceros)
    cer.append(saldocer)
    cel.append(saldocel)

dfc["SALDO PROPIA EXPOST"] = propia
dfc["SALDO TERCEROS EXPOST"] = terceros
dfc["SALDO CER EXPOST"] = cer
dfc["SALDO CEL EXPOST"] = cel

    
```
